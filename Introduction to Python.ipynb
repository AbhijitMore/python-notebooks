```python
characters = ['z', 'B', 'C', 'D']

# Iterate over each character
for char in characters:
    ascii_value = ord(char)
    binary_value = bin(ascii_value)[2:]  # Remove the '0b' prefix
    print(f"Binary representation of '{char}' is {binary_value}")
```

```python
sentence = "Hello, Sony!"

# Iterate over each character in the sentence
for char in sentence:
    ascii_value = ord(char)
    binary_value = bin(ascii_value)
    print(f"Binary representation of '{char}' is {binary_value}")
```

The binary representations for 'A', 'B', and 'C' are 01000001, 01000010, and 01000011, respectively. Each binary representation consists of 8 bits, with '0' representing a bit value of 0 and '1' representing a bit value of 1.

Nested while loop with an outer and inner loop

```python
a = int(input('Enter the number of loop iterations: '))
outer_counter = 0

while outer_counter < int(a):
    inner_counter = 0

    while inner_counter < 3:
        print("Outer loop:", outer_counter, "Inner loop:", inner_counter)
        inner_counter += 1

    outer_counter += 1
```

### Mutable and Immutable Sets

Although some of the mutable elements (like list) are not allowed in set, set itself is mutable.

```python
# Mutable Set
set6 = set(['A','B','C','D','E'])
set6.add('F')
print(set6)
```

```python
# Immutable Set
set7 = frozenset(['A','B','C'])
set7.add('D') # will generate error
set7
```

```python
# Sets are unordered, to get ordered set use sorted function

set8 = sorted(set6)
set8
```

```python
# Create a set called 'x' with elements 'a', 'b', 'c', 'd', and 'e'.
x = {"a", "b", "c", "d", "e"}

# Remove an arbitrary element from the set 'x' using the 'pop()' method.
x.pop()
```

```python
# Remove another arbitrary element from the set 'x'.
x.pop()
```

```python
# Difference between sets
x = {"a","b","c","d","e"}
y = {"b","c"}
z = {"c","d"}
x.difference(y)
```

```python
# Find the elements in 'x' that are not in 'y' using the 'difference()' method.
x.difference(y).difference(z)
```

```python
# instead of using the method difference, we can use the operator "-"
print(x - y)
print(x - y - z)
```

```python
# difference_update removes all elements of another set from this set.
x = {"a","b","c","d","e"}
y = {"b","c"}
x.difference_update(y)
print(x)
```

```python
# x.difference_update(y) is the same as "x = x - y"
x = {"a","b","c","d","e"}
y = {"b","c"}
x = x - y
print(x)
```
