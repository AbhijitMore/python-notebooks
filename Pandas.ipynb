# **PANDAS SERIES**



A Pandas Series is a one-dimensional labeled array that can hold data of any type. It is similar to a one-dimensional NumPy array but is associated with labels or indices, allowing for more flexible and meaningful data manipulation.




Key characteristics of a Pandas Series:

  1. **Homogeneous Data:** All elements in a Pandas Series must be of the same data type, unlike Python lists.

  2. **Labeled Index:** Each element in a Series has a label or index, which provides a way to access and manipulate the data.

  3. **Data Types:** A Pandas Series can hold data of any valid Python data type, such as integers, floats, strings, or even custom objects.

  4. **Operations:**  Series support various operations, including arithmetic operations, slicing, and statistical operations.







### Creating a Series:

````python
import pandas as pd


# From a list
s = pd.Series([1, 2, 3, 4, 5])

print(s,"\n")

# From a NumPy array
import numpy as np
s = pd.Series(np.array([1, 2, 3, 4, 5]))

print(s,"\n")

# With custom index
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])

print(s)
````

  Note : The default index always strarts rom zero and goes on till n-1 , where n is the  no. of elements in the particular series

## Operations:

````python
# Arithmetic operations
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
s_sum = s1 + s2
print(s_sum,"\n\n")

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([2, 3, 4])
s_sub = s1 - s2
print(s_sub,"\n\n")

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([2, 3, 4])
s_mult = s1*s2
print(s_mult,"\n\n")

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([2, 3, 4])
s_div = s1/s2
print(s_div)
````

### In the above code segment the respective elements get added , subtracted , multiplied , divided on index basis .

### Creating a Dataframe:

````python
# From a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)


print(df,"\n\n")
# From a list of dictionaries
data_list = [{'Name': 'Alice', 'Age': 25, 'City': 'New York'},
             {'Name': 'Bob', 'Age': 30, 'City': 'San Francisco'},
             {'Name': 'Charlie', 'Age': 35, 'City': 'Los Angeles'}]
df = pd.DataFrame(data_list)
print(df,"\n\n")

# From a NumPy array
df = pd.DataFrame(np.array([[1, 2, 3], [4, 5, 6]]), columns=['A', 'B', 'C'])
print(df,"\n\n")
````

## Basic Indexing

````python
import pandas as pd

# Creating a Series
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Accessing elements by label
print(s['a'])
# Accessing elements by position
print(s[0])

# Slicing
print(s['b':'d'])
````

````python
import pandas as pd

# Creating a Series with hierarchical index
index = [('A', 1), ('A', 2), ('B', 1), ('B', 2)]
values = [10, 20, 30, 40]
s = pd.Series(values, index=pd.MultiIndex.from_tuples(index, names=['letters', 'numbers']))

# Accessing elements
print(s['A'])
print(s['A', 1])
print(s.loc['A'])

````

## Boolean Indexing

````python
import pandas as pd

# Creating a Series
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Boolean indexing
print(s[s > 30])
````

## Fancy Indexing

````python
import pandas as pd

# Creating a Series
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

# Fancy indexing
print(s[['a', 'c', 'e']])
````

## Reindexing

````python
import pandas as pd

# Creating a Series
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# Reindexing with a new set of indices
s_reindexed = s.reindex(['a', 'b', 'c', 'd'], fill_value=400)
print(s_reindexed)
````

    The reindex method is used to change the index of the Series.
    In this example, we reindex the Series with a new set of indices ['a', 'b', 'c', 'd'].
    The fill_value parameter is set to 0, which fills missing values with the specified value.

## Interpolation

````python
import pandas as pd

# Creating a Series with missing values
s = pd.Series([10, None, 30, None, 50], index=['a', 'b', 'c', 'd', 'e'])

# Linear interpolation to fill missing values
s_interpolated = s.interpolate()
print(s_interpolated)
````

## Basic Indexing

````python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

# Accessing columns
print(df['Name'])

# Accessing rows by label
print(df.loc['a'])

# Accessing rows by position
print(df.iloc[0])
````

````python
import pandas as pd

# Creating a DataFrame with hierarchical index
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=['letters', 'numbers'])
df = pd.DataFrame({'Values': [10, 20, 30, 40]}, index=index)

# Accessing elements
print(df.loc['A'])


print(df.loc['A', 1])
print(df.loc['B', 2])
````

````python
import pandas as pd

# Creating a DataFrame with hierarchical index
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=['letters', 'numbers'])
df = pd.DataFrame({'Values': [10, 20, 30, 40]}, index=index)

# Accessing elements
print(df.loc['A'])
print(df.loc['A', 1])    # Output: Values    10
print(df.loc['B', 2])    # Output: Values    40

# Slicing with hierarchical index
print(df.loc['A':'B'])
````

**Cross-Section with xs:**

````python
import pandas as pd

# Creating a DataFrame with hierarchical index
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=['letters', 'numbers'])
df = pd.DataFrame({'Values': [10, 20, 30, 40]}, index=index)

# Using cross-section to select data
selected_values = df.xs(key='A', level='letters')
print(selected_values)
````

    The xs method is used to select data at a particular level of a hierarchical index.
    In this example, it selects rows where the level 'letters' is equal to 'A'.

**Sorting by Index**

````python
import pandas as pd

# Creating a DataFrame with hierarchical index
arrays = [['B', 'A', 'A', 'B'], [2, 1, 2, 1]]
index = pd.MultiIndex.from_arrays(arrays, names=['letters', 'numbers'])
df = pd.DataFrame({'Values': [10, 20, 30, 40]}, index=index)

# Sorting by index
df_sorted = df.sort_index()
print(df_sorted)
````

Handling missing data in Pandas refers to the process of managing and addressing data points that are not available or are represented as missing values (usually as NaN or None). Missing data can occur for various reasons, such as errors during data collection, incomplete data, or intentional gaps.

Pandas provides several methods and functions for handling missing data, allowing users to make informed decisions on how to deal with these gaps in their datasets.

1. Checking for Missing Data:

You can use the isnull() or notna() methods to check for missing data in a Series or DataFrame.

````python
import pandas as pd

# Creating a DataFrame with missing data
data = {'Name': ['Alice', 'Bob', None, 'Charlie'],
        'Age': [25, 30, None, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles', None]}
df = pd.DataFrame(data)

# Check for missing data
print(df.isnull().sum())
# or
print(df.notna().sum())

print(df.isna().sum())
````

````python
import pandas as pd

# Creating a DataFrame with missing values
data = {'Name': ['Alice', 'Bob', 'Charlie', None],
        'Age': [25, None, 35, 40],
        'City': ['New York', 'San Francisco', None, 'Los Angeles']}
df = pd.DataFrame(data)

# Check for missing values
print(df.isna())
````

2. Removing Missing Data:

Use the dropna() method to remove rows or columns with missing data.

````python
# Remove rows with any missing data
df_cleaned_rows = df.dropna()

# Remove columns with any missing data
df_cleaned_columns = df.dropna(axis=1)

# Remove rows with missing data in specific columns
df_cleaned_specific_columns = df.dropna(subset=['Name', 'Age'])
print(df,"\n\n")
print(df_cleaned_specific_columns)
````

3. Filling Missing Data:

You can use the fillna() method to fill missing values with a specified value or a calculated value.

````python
# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)

# Fill missing values with the mean of the column
df_filled_mean = df.fillna(df.mean())

# Forward-fill missing values (use the previous non-null value)
df_ffill = df.fillna(method='ffill')

# Backward-fill missing values (use the next non-null value)
df_bfill = df.fillna(method='bfill')


print(df_filled,"\n\n")
print(df_filled_mean,"\n\n")
print(df_ffill,"\n\n")
print(df_bfill)
````

4. Interpolation:

Use the interpolate() method to fill missing values with interpolated values.

````python
# Linear interpolation
df_interpolated = df.interpolate()

print(df_interpolated)
````

# Boolean indexing

Boolean indexing in Pandas involves using boolean expressions (conditions) to filter and select data in a Pandas Series or DataFrame. It entails creating a boolean mask, which is a series of True and False values based on whether each element in the original Series or DataFrame satisfies a given condition.

This technique follows a straightforward process:

    Boolean Condition:
        A boolean condition is formed by applying a comparison or logical operation to a Pandas Series or DataFrame, resulting in a boolean Series or DataFrame.

    Boolean Mask:
        The boolean condition serves as a mask, generating a boolean Series or DataFrame with True values where the condition is met and False values otherwise.

    Applying the Mask:
        The boolean mask is applied to select the desired subset of data from the original Series or DataFrame.

Boolean indexing is beneficial in various scenarios:

    Filtering Data:
        This method is powerful for extracting subsets of data that meet specific criteria, such as selecting rows with grades above a certain threshold.

    Combining Conditions:
        Multiple conditions can be combined using logical operators (and, or, not) to create complex queries for data selection.

    Missing Data Handling:
        Boolean indexing is useful for handling missing data by selecting or excluding rows with missing values, aiding in data cleaning.

    Conditional Assignment:
        Boolean indexing facilitates conditional assignment of values based on specific conditions, enabling efficient data transformation.

Boolean indexing is a flexible and expressive approach to perform data selection and manipulation in Pandas, widely used in data analysis for cleaning, transforming, and exploring datasets.

# Handling Outliers

Handling outliers refers to the process of identifying and addressing data points that deviate significantly from the overall pattern or distribution of the dataset. Outliers are values that are unusually high or low compared to the majority of the data and can adversely affect statistical analyses and machine learning models. Handling outliers is an essential step in data preprocessing to ensure the accuracy and reliability of analytical results. There are several common approaches to handling outliers:



  **Detection:**

        The first step is to detect outliers in the dataset.
        This can be done using statistical methods such as Z-Score or the Interquartile Range (IQR).
        Visualization tools like box plots or scatter plots can also be used for outlier detection.

  **Removal:**

        Outliers can be removed from the dataset if they are determined to be errors or anomalies.
        This involves excluding data points that fall outside a predefined range or threshold based on statistical measures.

  **Transformation:**

         Transforming the data can help reduce the impact of outliers.
         Common transformations include taking the logarithm or
         square root of the data, making the distribution more symmetrical and less sensitive to extreme values.

 ** **Imputation**:**

         In some cases, outliers are replaced or imputed with more typical values.
        This might involve replacing outliers with the mean, median, or a value predicted from the rest of the data.

 **Winsorizing**:
        
        Winsorizing involves capping extreme values by replacing them with values at a specified percentile.
        This method retains the overall distribution while mitigating the impact of outliers.

  **Binning:**

        Grouping data into bins can help minimize the influence of outliers. This is often done for visualization or summary purposes, where the focus is on ranges rather than individual data points.

  

### Handling Outliers in a Pandas Series:

1. **Visual Inspection:**
   - Use visualizations such as box plots or histograms to identify outliers in the data.

2. **Z-Score Method:**
   - Calculate Z-scores and filter out data points beyond a specified threshold.

    ```python
    from scipy.stats import zscore
    filtered_series = series[(zscore(series) < 3) & (zscore(series) > -3)]
    ```

3. **Percentile Method (IQR):**
   - Use the Interquartile Range (IQR) to filter out values outside a specified percentile range.

    ```python
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    filtered_series = series[(series >= q1 - 1.5 * iqr) & (series <= q3 + 1.5 * iqr)]
    ```



Z-Score provides information about the standardized position of a data point in relation to the mean and standard deviation.
IQR is a robust measure that characterizes the spread of the central portion of the dataset and is effective in identifying outliers, especially in non-normally distributed data.

What is Data Normalisation ?

Data normalization, also known as data scaling or feature scaling, is the process of transforming numerical data into a standardized format. The goal of normalization is to bring the data into a common scale, making it easier to compare and analyze. This is particularly important when working with machine learning algorithms, as many of them are sensitive to the scale of input features. The normalization process typically involves scaling the data to a specific range or distribution.

Common methods for data normalization include:

    Min-Max Scaling   
    Z-Score Standardization
    Robust Scaling
    Decimal Scaling

Data normalization is crucial for the following reasons:

    Equalizing Scales: Different features often have different scales, and normalizing them ensures that they contribute equally to the analysis.

    Facilitating Convergence: Many machine learning algorithms, such as gradient-based optimization methods, converge faster when input features are on a similar scale.

    Improving Interpretability: Normalized data is easier to interpret since the scale of each feature is consistent.

    Handling Outliers: Normalization can reduce the impact of outliers on the analysis.

What is transformation of data ?

In the context of data analysis and statistics, transformation of data refers to the process of applying a function or operation to the original data with the goal of altering its structure, distribution, or characteristics. The purpose of data transformation varies depending on the specific objectives of the analysis and the properties of the data. Here are some common types of data transformations and their purposes:

1. **Normalization:**
   - **Purpose:** To scale numerical values to a standard range, often between 0 and 1, making them comparable and reducing the impact of differences in scale.
   - **Example:** Min-Max Scaling, Z-Score Standardization.

2. **Logarithmic Transformation:**
   - **Purpose:** To reduce the impact of outliers and make the data conform more closely to a normal distribution.
   - **Example:** Taking the natural logarithm of values.

3. **Exponential Transformation:**
   - **Purpose:** To emphasize or de-emphasize certain patterns in the data, especially when dealing with growth rates or trends.
   - **Example:** Raising values to a power, such as squaring or cubing.

4. **Box-Cox Transformation:**
   - **Purpose:** Similar to logarithmic transformations, it stabilizes the variance and makes the data more closely approximate a normal distribution.
   - **Example:** Box-Cox power transformation.

5. **Standardization:**
   - **Purpose:** To center the data by subtracting the mean and dividing by the standard deviation, resulting in a distribution with a mean of 0 and standard deviation of 1.
   - **Example:** Z-Score Standardization.

6. **Categorical Encoding:**
   - **Purpose:** To represent categorical variables as numerical values, enabling their use in mathematical models.
   - **Example:** One-Hot Encoding, Label Encoding.

7. **Grouped Transformations:**
   - **Purpose:** To apply transformations within groups of data, often used in combination with the `groupby` operation.
   - **Example:** Group-wise standardization or normalization.

8. **Scaling:**
   - **Purpose:** To scale numerical values proportionally, maintaining the relative distances between data points.
   - **Example:** Min-Max Scaling, Robust Scaling.

Data transformation is a crucial step in data preprocessing, helping to address issues such as outliers, skewed distributions, and scale differences. The choice of transformation depends on the specific characteristics of the data and the requirements of the analysis or modeling process.

What is filtering of data ?

Filtering of data refers to the process of selecting a subset of data from a larger dataset based on specific conditions or criteria. In other words, filtering allows you to focus on and extract the rows or columns of a dataset that meet certain requirements or satisfy particular conditions. This is a fundamental operation in data analysis and is often used to isolate and examine specific aspects of the data.

Commonly, filtering is performed using logical conditions applied to the values in one or more columns of the dataset. The resulting subset of data retains only those rows that fulfill the specified conditions.

Here are some key aspects of data filtering:

    Filtering Rows:
        Selecting rows based on conditions such as numerical comparisons, text matching, or a combination of criteria.
        Example: Selecting all rows where the value in the 'Age' column is greater than 30.

    Filtering Columns:
        Selecting specific columns of interest while excluding others.
        Example: Selecting only the 'Name' and 'Salary' columns from a dataset.

    Combining Conditions:
        Using logical operators (AND, OR) to create more complex filtering conditions.
        Example: Selecting rows where both 'Age' is greater than 30 and 'Salary' is above a certain threshold.

    Boolean Indexing:
        Creating boolean masks or conditions and using them to index into the DataFrame.
        Example: Creating a boolean mask for rows with 'Category' equal to 'A' and using it to filter the DataFrame.

    Querying:
        Using the query() method to filter data based on a query string.
        Example: Querying the DataFrame to select rows where 'Quantity' is greater than 10.

# WHAT IS TIME SERIES DATA ?

Time series data is a type of data where the values are associated with specific points in time and are ordered chronologically. In a time series, observations or measurements are recorded at successive and equally spaced intervals. This data structure is particularly relevant for analyzing and understanding patterns, trends, and behaviors that evolve over time.

Key characteristics of time series data include:

    Temporal Order:
        The data points are arranged in a chronological sequence, with each observation associated with a specific time or time period.

    Equally Spaced Intervals:
        The time intervals between consecutive data points are uniform, meaning that the time intervals are consistent and regular.

    Dependencies:
        There is often a temporal dependence between observations. The value at a given time is influenced by the values at earlier times.

    Patterns and Trends:
        Time series data often exhibits patterns, trends, and seasonality. Analyzing these patterns can provide insights into underlying processes or phenomena.

Examples of time series data include:

    Stock Prices: Daily closing prices of a stock over time.

    Weather Data: Daily temperature, humidity, or precipitation measurements over a period.

    Economic Indicators: Monthly unemployment rates, GDP growth, or inflation rates.

    Web Traffic: Hourly or daily website visit counts.

    Sensor Readings: Time-stamped measurements from sensors, such as temperature sensors, heart rate monitors, etc.

Analyzing time series data involves techniques such as:

    Trend Analysis: Identifying overall trends or long-term movements in the data.

    Seasonal Decomposition: Separating a time series into components like trend, seasonality, and residual (error) to better understand its structure.

    Forecasting: Predicting future values based on historical data.

    Smoothing Techniques: Reducing noise and fluctuations in the data to reveal underlying patterns.

Time series analysis is widely used in various domains, including finance, economics, environmental science, healthcare, and engineering, to make informed decisions, predictions, and understand dynamic phenomena over time.

The resulting dataframe will contain all unique keys from both df1 and df2, with NaN values where a key is not present in one of the dataframes.

    Description: Keeps all rows from the left dataframe and fills in missing values with NaN for the right dataframe.

````python
merged_left = pd.merge(df1, df2, on='Key', how='left')
print(merged_left)
````

The resulting dataframe will contain all rows from df1 and include matching rows from df2, with NaN values for columns from df2 where there is no match.

    Description: Keeps all rows from the right dataframe and fills in missing values with NaN for the left dataframe.

````python
merged_right = pd.merge(df1, df2, on='Key', how='right')
print(merged_right)
````

The resulting dataframe will contain all rows from df2 and include matching rows from df1, with NaN values for columns from df1 where there is no match.

In pandas, joining dataframes involves combining two or more dataframes based on a common column or index. There are several types of joins, and each has different behaviors.



### 1. Inner Join:

An inner join returns only the rows where there is a match in both dataframes based on the specified key.

````python
import pandas as pd

# Create two dataframes
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['B', 'C', 'D'], 'Value': [4, 5, 6]})

# Inner join on the 'Key' column
result_inner = pd.merge(df1, df2, on='Key', how='inner')

# Display the result
print(result_inner)
````

### 2. Left Join:

A left join returns all the rows from the left dataframe and the matched rows from the right dataframe. If there is no match, NaN values are filled for the columns from the right dataframe.



````python
# Left join on the 'Key' column
result_left = pd.merge(df1, df2, on='Key', how='left')

# Display the result
print(result_left)
````

### 3. Right Join:

A right join is similar to a left join but returns all the rows from the right dataframe.



````python
# Right join on the 'Key' column
result_right = pd.merge(df1, df2, on='Key', how='right')

# Display the result
print(result_right)
````

### 4. Outer Join:

An outer join returns all rows when there is a match in either the left or right dataframe. NaN values are filled for columns where there is no match.


In this example, all rows from both dataframes are included, and NaN values are filled for columns where there is no match.

These are the basic types of joins in pandas, and you can choose the appropriate join based on your specific data merging requirements. The `on` parameter specifies the key column, and the `how` parameter determines the type of join.

````python
# Outer join on the 'Key' column
result_outer = pd.merge(df1, df2, on='Key', how='outer')

# Display the result
print(result_outer)
````

````python
import pandas as pd

# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 15, 25]
}

df = pd.DataFrame(data)

# Create a basic pivot table
pivot_table = pd.pivot_table(df, values='Value', index='Date', columns='Category', aggfunc='sum')

# Display the result
print(pivot_table)
````

Categorical data can be divided into several types based on the nature of the categories and their relationships. The main types of categorical data include:

### 1. **Nominal Data:**
   - **Definition:** Nominal data represents categories with no inherent order or ranking.
   - **Example:** Colors (e.g., red, blue, green), Gender (e.g., male, female), Types of fruits (e.g., apple, banana, orange).
   - **Note:** Nominal data doesn't have a meaningful order, and mathematical operations like addition or subtraction are not meaningful.

### 2. **Ordinal Data:**
   - **Definition:** Ordinal data represents categories with a clear order or ranking but the intervals between them may not be consistent.
   - **Example:** Educational levels (e.g., high school, college, graduate), Customer satisfaction levels (e.g., low, medium, high), Rank in a competition (e.g., 1st, 2nd, 3rd).
   - **Note:** Ordinal data allows for meaningful comparisons of the order but not necessarily the intervals between categories.

### 3. **Binary Data:**
   - **Definition:** Binary data represents only two categories.
   - **Example:** Yes/No, True/False, Success/Failure.
   - **Note:** Binary data is a special case of nominal data with exactly two categories.

### 4. **Multi-Categorical Data:**
   - **Definition:** Multi-categorical data represents more than two categories.
   - **Example:** Types of cars (e.g., sedan, SUV, truck), Days of the week (e.g., Monday, Tuesday, Wednesday).
   - **Note:** Multi-categorical data can be nominal or ordinal depending on the nature of the categories.

### 5. **Derived Categorical Data:**
   - **Definition:** Derived categorical data is created by grouping or transforming existing numerical data into categories.
   - **Example:** Age groups (e.g., 0-18, 19-30, 31-45), Income brackets (e.g., low, medium, high).
   - **Note:** Derived categorical data is often ordinal and created for simplifying analysis or improving interpretability.

### 6. **Count Data:**
   - **Definition:** Count data represents the frequency or occurrence of different categories.
   - **Example:** Number of emails received per day (0, 1, 2, ...), Number of defects in a product (0, 1, 2, ...).
   - **Note:** Count data is often treated as nominal or ordinal based on the nature of the categories.

Understanding the type of categorical data is important for choosing appropriate statistical analyses, visualizations, and modeling techniques. Different types of categorical data may require different handling in data preprocessing and analysis.

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a DataFrame with different types of categorical data
data = {
    'Nominal': ['Red', 'Blue', 'Green', 'Red', 'Green'],
    'Ordinal': ['Low', 'Medium', 'High', 'Medium', 'Low'],
    'Binary': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'MultiCategorical': ['Sedan', 'SUV', 'Truck', 'Sedan', 'SUV'],
    'Age': [25, 35, 45, 28, 40]
}

df = pd.DataFrame(data)

# Convert 'Nominal' and 'Ordinal' columns to categorical type
df['Nominal'] = df['Nominal'].astype('category')
df['Ordinal'] = pd.Categorical(df['Ordinal'], categories=['Low', 'Medium', 'High'], ordered=True)

# Create a derived categorical column 'AgeGroup'
bins = [0, 30, 40, 50, np.inf]
labels = ['0-30', '31-40', '41-50', '51+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Plot bar charts for each categorical column
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

df['Nominal'].value_counts().plot(kind='bar', ax=axes[0, 0], color='skyblue', edgecolor='black')
axes[0, 0].set_title('Nominal Data')

df['Ordinal'].value_counts().plot(kind='bar', ax=axes[0, 1], color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Ordinal Data')

df['Binary'].value_counts().plot(kind='bar', ax=axes[0, 2], color='salmon', edgecolor='black')
axes[0, 2].set_title('Binary Data')

df['MultiCategorical'].value_counts().plot(kind='bar', ax=axes[1, 0], color='gold', edgecolor='black')
axes[1, 0].set_title('Multi-Categorical Data')

df['AgeGroup'].value_counts().plot(kind='bar', ax=axes[1, 1], color='lightcoral', edgecolor='black')
axes[1, 1].set_title('Derived Categorical Data (Age Groups)')

# Hide the empty subplot
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
````

Handling categorical variables often involves converting them into a numerical format that can be used in machine learning models. Two common techniques for encoding categorical variables are Label Encoding and One-Hot Encoding. Let's go through explanations and coding snippets for both.

### 1. Label Encoding:

Label Encoding involves assigning a unique integer to each category in a categorical variable. This method is suitable for ordinal data where there is a meaningful order among the categories.


````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Creating a sample DataFrame
data = {'Category': ['Low', 'Medium', 'High', 'Low', 'Medium']}
df = pd.DataFrame(data)

# Initializing the LabelEncoder
label_encoder = LabelEncoder()

# Applying Label Encoding to the 'Category' column
df['Category_LabelEncoded'] = label_encoder.fit_transform(df['Category'])

# Display the DataFrame
print("Original DataFrame:")
print(df[['Category']])

# Display the DataFrame with Label Encoded column
print("\nDataFrame with Label Encoded Column:")
print(df[['Category', 'Category_LabelEncoded']])


````

### 2. One-Hot Encoding:

One-Hot Encoding is suitable for nominal data. It creates binary columns for each category and indicates the presence or absence of a category with a 1 or 0.



````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Creating a sample DataFrame
data = {'Category': ['Red', 'Blue', 'Green', 'Red', 'Green']}
df = pd.DataFrame(data)

# Initializing the OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)

# Applying One-Hot Encoding to the 'Category' column
encoded_categories = onehot_encoder.fit_transform(df[['Category']])
encoded_df = pd.DataFrame(encoded_categories, columns=onehot_encoder.get_feature_names_out(['Category']))

# Concatenating the original DataFrame and the one-hot encoded DataFrame
df_encoded = pd.concat([df, encoded_df], axis=1)

# Display the original DataFrame
print("Original DataFrame:")
print(df[['Category']])

# Display the DataFrame with One-Hot Encoded columns
print("\nDataFrame with One-Hot Encoded Columns:")
print(df_encoded)
````

What is vectorization ?

Vectorization is a technique used in pandas and NumPy to apply operations on entire arrays or columns of data at once, rather than iterating through the elements one by one. This can significantly improve data performance by taking advantage of optimized, low-level code and parallelization.


A vector database is designed to store, manage and index massive quantities of high-dimensional vector data efficiently. These databases are rapidly growing in interest to create additional value for generative artificial intelligence (AI) use cases and applications.

Vector databases are typically used to power vector search use cases like visual, semantic, and multimodal search. More recently, they're paired with generative artificial intelligence (AI) text models to create intelligent agents that provide conversational search experiences.

How to use it improve performance ?

Vectorization is a technique used in computer programming to express operations as vector operations, which can lead to significant performance improvements. In the context of data processing, vectorization is commonly associated with numerical and scientific computing, where large datasets can be processed more efficiently using vectorized operations.

Here are the general steps to use vectorization and how it can improve performance:

### 1. **Use Vectorized Libraries:**
   - Utilize libraries that are optimized for vectorized operations. NumPy is a popular library for numerical operations in Python and is designed to perform array operations efficiently.

### 2. **Avoid Loops:**
   - Replace explicit loops with vectorized operations whenever possible. Loops in interpreted languages like Python can be slow, and vectorized operations leverage optimized, compiled code to perform operations on entire arrays or matrices at once.

### 3. **Use Array Operations:**
   - Express operations as array operations. Instead of iterating through each element in an array, apply the operation directly to the entire array or a subset of it.

### 4. **Broadcasting:**
   - Understand and leverage broadcasting. Broadcasting allows operations between arrays of different shapes and sizes to be executed efficiently without the need for explicit loop iteration.



````python
#Example in NumPy (Python):
import numpy as np
a=[1,2,3,4,5,6,7,8,9]
b=[9,8,7,6,5,4,3,2,1]
   # Non-vectorized approach
def non_vectorized_add(a, b):
  result = []
  for i in range(len(a)):
    result.append(a[i] + b[i])
    return result

   # Vectorized approach
def vectorized_add(a, b):
  return np.array(a) + np.array(b)

print("Non-vectorized approach")
print(non_vectorized_add(a,b),"\n\n")
print("Vectorized approach")
print(vectorized_add(a,b))


````

### 5. **Performance Benefits:**
   - Vectorization takes advantage of hardware-level optimizations and parallelization, making it much faster than explicit looping.
   - It leads to cleaner and more concise code, improving code readability and maintainability.
   - The use of optimized libraries (like NumPy) further enhances performance.

### 6. **Considerations:**
   - While vectorization is powerful, it may not always be applicable or beneficial in all situations. It depends on the nature of the operations and the size of the data.
   - Vectorized operations are particularly effective when dealing with large datasets or complex mathematical operations.

### 7. **Parallelization:**
   - Modern hardware, such as multicore processors and GPUs, can further enhance performance through parallelization. Some vectorized operations can be automatically parallelized by the underlying libraries.

### 8. **Measure Performance:**
   - Always measure the performance of your code using profiling tools to identify bottlenecks and optimize accordingly.

In summary, vectorization is a powerful technique to improve the performance of data processing code, especially in numerical and scientific computing. It allows you to perform operations on entire arrays or matrices at once, leveraging optimized libraries and hardware capabilities.

## Example 1: Element-wise Operations

````python
import numpy as np

# Element-wise addition
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
result = arr1 + arr2
print(result)

# Element-wise multiplication
result = arr1 * arr2
print(result)

# Element-wise square root
result = np.sqrt(arr1)
print(result)
````

## Example 2: Broadcasting

````python
# Broadcasting scalar to array
arr = np.array([1, 2, 3, 4])
result = arr + 10
print(result)

# Broadcasting with different-sized arrays
matrix = np.array([[1, 2, 3], [4, 5, 6]])
row_vector = np.array([10, 20, 30])
result = matrix + row_vector
print(result)

````

## Example 3: Mathematical Functions

````python
# Applying mathematical functions to arrays
arr = np.array([1, 2, 3, 4])
result = np.exp(arr)
print(result)  # Output: [2.718, 7.389, 20.085, 54.598]

# Trigonometric functions
result_sin = np.sin(arr)
result_cos = np.cos(arr)
print(result_sin)
print(result_cos)
````

## Example 4: Logical Operations

````python
# Logical operations on arrays
arr = np.array([True, False, True, False])
result_and = arr & (arr == True)
result_or = arr | (arr == False)
print(result_and)
print(result_or)
````

## Example 5: Scalar Operations:

````python
import numpy as np

# Non-vectorized approach
a = [1, 2, 3, 4]
scalar = 2
result = [ai * scalar for ai in a]

# Vectorized approach
a_array = np.array(a)
result_vectorized = a_array * scalar

print(result_vectorized)
````

## Example 6: Matrix Multiplication:

````python
import numpy as np

# Non-vectorized approach
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
result = np.matmul(A, B)

# Vectorized approach
A_array = np.array(A)
B_array = np.array(B)
result_vectorized = A_array @ B_array

print(result_vectorized)
````

## Example 7: Conditional Operations:

````python
import numpy as np

# Non-vectorized approach
a = np.array([1, 2, 3, 4, 5])
result = [ai if ai > 2 else 0 for ai in a]

# Vectorized approach
result_vectorized = np.where(a > 2, a, 0)

print(result_vectorized)
````

Vectorized operations leverage optimized, low-level, and parallelized routines, enabling efficient processing of entire arrays in a single operation. This reduces the need for explicit loops, enhancing computational efficiency by harnessing hardware capabilities and minimizing interpretive overhead when processing large dataframes of data .
